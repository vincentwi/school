# Segmentation

```python
#Creating aggregates per clients 
df_contact_sum = df.drop(["Batch number", "Receipt date", "cam code", "Operation", "Segment", "Movement Id", "Ventilation Id", "oft code", "payment method", "RF"], axis = 1).groupby(["contact ID"]).sum()
df_contact_sum.shape
```

```python
#Creating aggregates per clients 
df_contact_avg = df.drop(["Batch number", "Receipt date", "cam code", "Operation", "Segment", "Movement Id", "Ventilation Id", "oft code", "payment method", "RF"], axis = 1).groupby(["contact ID"]).mean()
df_contact_avg.shape
```

```python
#Creating aggregates per clients 
df_contact_max = df.drop(["Batch number", "cam code", "Operation", "Segment", "Movement Id", "Ventilation Id", "oft code", "payment method", "RF"], axis = 1).groupby(["contact ID"]).max()
df_contact_max.shape
```

```python
df_contact = df_contact_sum.merge(df_contact_avg, left_on = "contact ID", right_on = "contact ID", suffixes=('_sum', '_avg')).merge(df_contact_max, left_on = "contact ID", right_on = "contact ID", suffixes=('_', '_max'))
df_contact.columns = ["total_transaction_amount", "total_ventilation_amount", "mean_transaction_amount", "mean_ventilation_amount", "latest_donation", "max_transaction_amount", "max_ventilation_amount"]
df_contact.head()
```

```python
#Convert latest transaction to recency in days
df_contact["current_date"] = "2019-01-01"
df_contact["current_date"] = pd.to_datetime(df_contact["current_date"])
df_contact["recency"] = (df_contact["current_date"] - df_contact["latest_donation"]).dt.days
df_contact = df_contact.drop(["current_date", "latest_donation"], axis = 1)

df_contact = df_contact.loc[df_contact["mean_transaction_amount"].isna() == False]
df_contact.head()
```

```python
#Clustering
from sklearn.cluster import KMeans

kmeans = KMeans(5)
kmeans.fit(df_contact.to_numpy())
kmeans.cluster_centers_
```

```python
client_id = [1,2,3,4,5,6]
client_segment_2016 = [0,3,2,1,4,0]
client_segment_2017 = [0,2,3,0,3,1]

df = pd.DataFrame({"client_id": client_id, "client_segment_2016": client_segment_2016, "client_segment_2017": client_segment_2017})
df
```

```python
reference_count = df["client_segment_2016"].value_counts()
reference_count[2]
```

```python
df["client_segment_2017"].loc[df["client_segment_2016"] == 0].value_counts()
```

```python
import numpy as np

reference_count = df["client_segment_2016"].value_counts()
heatmap = np.zeros((5,5))
for i in range(5):
    count_2016 = reference_count[i]
    count_2017 = df["client_segment_2017"].loc[df["client_segment_2016"] == i].value_counts()
    for j in count_2017.index:
        heatmap[i,j] = count_2017[j] * 100 / count_2016 
print(heatmap)
```

```python
plt.imshow(heatmap, cmap='hot', interpolation='nearest')
```
