```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
alpha = 0.99
beta = 1
sigma = 0.09
```

```python
x = []
x0 = np.random.normal(0, np.sqrt(sigma**2 / (1-alpha**2)))
x.append(x0)
for i in range(1000):
  val = np.random.normal(x[-1]*alpha, sigma)
  x.append(val)


x = np.array(np.exp(x[1:]))
x[:4]
```

```python
plt.plot(x)
```

```python
y = []
for i in range(1, 1000):
  x_i = x[i]
  y_i = np.random.normal(0, np.sqrt(beta**2 * x_i))
  y.append(y_i)
```

```python
plt.plot(x, label="X")
plt.plot(y, alpha = 0.2, label="Y")
plt.legend()
plt.show()
```
