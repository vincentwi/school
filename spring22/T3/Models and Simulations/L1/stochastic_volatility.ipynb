# Exercise 1: stochastic volatility model

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# number of days
T = 1000
```

```python
# model parameters 
beta = 1.0 # variance of the returns
alpha = 0.99 # autoregressive parameter 
sigma = 0.1 # variance of the volatility process
```

```python
# simulate states and observations
X = np.zeros(T+1) # store latent states
Y = np.zeros(T+1) # store the observations

# initialization 
X[0] = np.random.normal(0, sigma / np.sqrt(1 - alpha**2)) # initial state
Y[0] = np.random.normal(0, beta * np.sqrt(np.exp(X[0]))) # initial observation

for t in range(T): # forward in time
    # simulate next latent volatility 
    X[t+1] = np.random.normal(alpha * X[t], sigma) 
    
    # simulate next observation
    Y[t+1] = np.random.normal(0, beta * np.sqrt(np.exp(X[t+1]))) 
```

```python
# plot time series
time = np.arange(0, T+1)
plt.figure()
plt.subplot(2,1,1)
plt.plot(time, np.exp(X)) 
plt.ylabel('Volatility',fontsize = 15)
plt.subplot(2,1,2)
plt.plot(time, Y)
plt.ylabel('Returns',fontsize = 15)
```
