```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# rate parameter
Lambda = 2

# sample size
N = 100000

# sample uniforms
U = np.random.rand(N)

# inverse CDF
X = - np.log(1 - U) / Lambda
```

```python
# plot histogram of samples 
plt.hist(X, bins = 100, density = True)

# overlay probability density function 
xgrid = np.linspace(min(X), max(X), 100) # create a grid of values from minimum sample to maximum sample
pdf = Lambda * np.exp(- Lambda * xgrid) # evaluate exponential probability density function 
plt.plot(xgrid, pdf) 
plt.show()

# histogram should agree with probability density function!
```
